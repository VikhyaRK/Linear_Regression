```python
# Linear Regression for y = 3x + 5
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)


x = [np.random.randint(1, 500) for _ in range(100)]
y4 = []


for i in x:
    y_pred = (3 * i) + 5 + (np.random.normal(0, 0.2))
    y4.append(y_pred)


w = [0, 0]
n = len(x)
learn_rate = 0.0001
D0plot = []
D1plot = []


for i in range(100000):
    D1 = 0
    D0 = 0
    for j in range(n):
        y_pred = w[0] * x[j] + w[1]
        y_real = y4[j]
        y_diff = y_pred - y_real
        D0 = (2 * y_diff * x[j]) / n
        D1 = (2 * y_diff) / n
    D0plot.append(D0)
    D1plot.append(D1)
    w[1] = w[1] - learn_rate * D1
    w[0] = w[0] - learn_rate * D0

print("Weights for y = 3x + 5:", w)


y_pred = [w[0] * i + w[1] for i in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y_pred, color='green', label='Regression Line (y = 3x + 5)')
plt.scatter(x, y4, color='purple', label='Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression: y = 3x + 5')
plt.show()
```
